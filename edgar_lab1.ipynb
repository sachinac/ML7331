One Hot Encoding of Categorical Variables

```python
# The data in this attribute is presented as a bynary option 0 and Y.
# To facilitate visualizing this attribute we used the get_dummies 
# variable to One Hot Encode the categorical vatiable RenalDiseaseIndicator.

one_hot_RenalDisease = pd.get_dummies(resultib_imputed['RenalDiseaseIndicator'],prefix='RenalDisease')

one_hot_RenalDisease.head()
```

```python
# Enable plot in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter('ignore', DeprecationWarning)

plt.style.use('ggplot')
one_hot_RenalDisease_grouped = one_hot_RenalDisease.groupby(by=['RenalDisease_0','RenalDisease_Y'])
#ax = one_hot_RenalDisease.plot(kind='barh')
one_hot_RenalDisease.head()
```

In 2009 out of the 41783 Medicare claims more than 30,000 had a negative Renal Disease Indicator, and aproximatelly 10,000 had a positive Renal Disease Indicator. By One Hot Encoding the RenalDiseaseIndicator attribute, we were able to graphycally represent these values using a numeric representation for negative and positive indicators, instead of the original mix of 0 and Y.

```python
plot_onehot = one_hot_RenalDisease[['RenalDisease_0','RenalDisease_Y']]

ax= sns.countplot(x='RenalDisease_Y', data = plot_onehot)    
plt.title('Beneficiary with Renal Disease Indicator')
plt.xticks(rotation=90)
plt.show()

plt.savefig('BeneficiatywithgRenalDiseaseIndicator')
```
